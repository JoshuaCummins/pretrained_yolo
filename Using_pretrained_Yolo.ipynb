#download yolo or tiny yolo.

```
#Code for video detection taking straight from the documentation.
detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
#set the yolo h5 file to the path in your google drive.
detector.setModelPath( os.path.join(execution_path , "/content/drive/MyDrive/3 class mask detection/pretrained-yolov3.h5"))
detector.loadModel()
#set the video file you have in your google drive to the path and output is the file name you want
video_path = detector.detectObjectsFromVideo(input_file_path=os.path.join(execution_path, "/content/drive/MyDrive/videos/comiccon.mp4"),
                            output_file_path=os.path.join(execution_path, "comiccon")
                            , frames_per_second=30, log_progress=True)
print(video_path)
```

```
# For images. You should use regular YOLO and not tiny since it won't take a long time to compute.
from imageai.Detection import ObjectDetection

execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "/content/drive/MyDrive/3 class mask detection/pretrained-yolov3.h5"))
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , "/content/drive/MyDrive/pics/selfie.jpg"), output_image_path=os.path.join(execution_path , "imagenew.jpg"), minimum_percentage_probability=30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```
